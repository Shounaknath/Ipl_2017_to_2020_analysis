```python
#Loading  the required libary
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loding the ipl matches dataset.
ipl=pd.read_csv("matches.csv")
```

```python
# Wanna know how many column and row is dataset..
ipl.shape
```

```python
ipl.tail()
```

```python
#Getting the frequency of  most  man of the match
ipl["player_of_match"].value_counts()
```

```python
#Getting the top 10 player with most man of the match
ipl["player_of_match"].value_counts()[0:10]
```

```python
#Getting the top 5 player with most man of the match
ipl["player_of_match"].value_counts()[0:5]
```

```python
list(ipl["player_of_match"].value_counts()[0:5].keys())
```

```python
#Making a bar plot for the top 5 players with most man of the match...

plt.figure(figsize=(9,5))
plt.bar(list(ipl["player_of_match"].value_counts()[0:5].keys()) , list(ipl["player_of_match"].value_counts()[0:5]))
plt.show()
```

```python
#Getting the frequency  of result column..
ipl["result"].value_counts()
```

```python
# Finding out the number of toss wins w.r.t each team...

ipl['toss_winner'].value_counts()
```

```python
# Extracting the records where a team  won batting first..
batting_first = ipl[ipl['win_by_runs']!=0]
```

```python
batting_first["winner"].value_counts()[0:5].keys()
```

```python
#Making a bar plot for the top 5 Team with most batting first win the match...
plt.figure(figsize=(7,5))
plt.bar(list(batting_first['winner'].value_counts()[0:4].keys()),list(batting_first['winner'].value_counts()[0:4]),color=["blue","yellow","red","purple"])
plt.show()
```

```python
# making piechart...
plt.figure(figsize=(9,9))
plt.pie(list(batting_first['winner'].value_counts()), labels=list(batting_first['winner'].value_counts().keys()), autopct='%0.1f%%')
plt.show()
```

```python
# makking a histogram ...
plt.figure(figsize=(7,5))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
# Extracting the records where a team  won batting Second..
batting_second=ipl[ipl["win_by_wickets"]!=0]
```

```python
batting_second.head()
```

```python
# Finding out the frequency of number of wins w.r.t each time after batting second..
batting_second["winner"].value_counts()
```

```python
#Taking a histogram for frequency of wins w.r.t  number of wickets.........
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
# Making a bar plot for top 3 teams with most wins after batting second..
plt.figure(figsize=(9,7))
plt.bar(list(batting_second['winner'].value_counts()[0:4].keys()),list(batting_second['winner'].value_counts()[0:4]), color=['purple','blue','red','yellow'])
plt.show()
```

```python
# Making Pie chart for distibution of most wins after batting second..
plt.figure(figsize=(7,8))
plt.pie(list(batting_second['winner'].value_counts()), labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
